```python
# Machine Learning!
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df_accident = pd.read_csv('UK_Traffic_Accidents_2015.csv')
df_accident.head()
```

```python
# drop columns
df_accident = df_accident.drop(columns=['Accident_Index','Location_Easting_OSGR','Location_Northing_OSGR','Date','Local_Authority_(Highway)','LSOA_of_Accident_Location'],axis=1)
df_accident.head()
```

```python
df_accident.fillna(value=0, inplace=True)
# df_accident.info()
df_accident['Time'] = df_accident['Time'].map(lambda x: str(x)[:-3])
df_accident['Time'] = df_accident['Time'].replace('',0)
df_accident['Time'] = df_accident['Time'].astype(int)
```

```python
# Build classification model
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix
from sklearn.preprocessing import MinMaxScaler
```

```python
def over(num):
    if num == 2:
        return 1
    elif num == 3:
        return 2
    else:
        return 0
df_accident['Accident_Severity'] = df_accident['Accident_Severity'].apply(over)
```

```python
scaler = MinMaxScaler()
x = df_accident.drop(columns=['Accident_Severity'])
x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
y = df_accident['Accident_Severity']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.35,random_state=0)
```

```python
clf = AdaBoostClassifier(n_estimators=100,random_state=0)
clf.fit(x_train,y_train)
```

```python
predict = clf.predict(x_test)
print('acu', accuracy_score(y_test,predict))
print('f1', f1_score(y_test,predict, average='micro'))
print('clas', classification_report(y_test,predict, zero_division=0))
```
